```python
import dlt
from pyspark.sql import functions as F
from pyspark.sql.functions import col, current_timestamp

path_city = "/Volumes/trips/landing/city"
schema_path = "/Volumes/trips/landing/city/_schema"  # ✅ Lives inside a valid volume

@dlt.table(
    name="trips.bronze.city",
    comment="Raw city data from CSV files",
    table_properties={
        "quality": "bronze"
    }
)
def raw_city_bronze():
    df = (
        spark.readStream
            .format("cloudFiles")
            .option("cloudFiles.format", "csv")
            .option("header", "true")
            .option("inferSchema", "true")
            .option("cloudFiles.schemaLocation", schema_path)  # ✅
            .load(path_city)
    )
    return df.withColumn("ingestion_date", F.current_timestamp()).withColumn("ingest_datetime", current_timestamp())
```
